```python
# filtering the teams data for HR>=200 for the Chicago cubs
cubs_HR = teams[(teams["teamID"] == "CHN") & (teams["HR"] >= 200)]

# finding the number of unique years the cubs had HR>=200
total_years = len(cubs_HR["yearID"].unique())
print("Number of years the Chicago Cubs hit at least 200 HRs:" ,total_years)

# finding the median of the wins
median_wins = cubs_HR["W"].median()
print("\nThe median number of wins in the seasons the Chicago Cubs hit at least 200 HRs:" ,median_wins)
```

## (c)

```python
# creating the factor election based on the presidential terms
teams["election"] = ((teams["yearID"]-1788) // 4) + 1
teams.head()
```

```python
# finding the home runs for each 4 year term
HR_term = teams.groupby('election')['HR'].sum().reset_index()
HR_term.head()
```

```python
# Finding the presidential term with the most home runs
HR_term_max = HR_term[HR_term['HR'] == HR_term['HR'].max()]
HR_term_max
```

### The presidential term with the most home runs is the 58th preseidential term with a total of 24076 home runs

## (a)

```python
# Filling the missing values in the grade column as Missing
violations['grade'] = violations['grade'].fillna('Missing')

# Calculating the count of inspections with a grade of 'A' and grouping by boro.
inspection_A = violations[violations['grade'] == 'A'].groupby('boro')['grade'].count()

# Calculating the total count of inspections of grade and grouping by boror
total_inspections = violations.groupby('boro')['grade'].count()

# Calculate the proportion
proportion_A = inspection_A / total_inspections

# Creating a DataFrame to display the column name Proportion_A
result_df = pd.DataFrame({'Proportion_A': proportion_A})

result_df
```

## (b)

```python
# Finding the top ten dba's with the most number of inspections
top_dba = violations['dba'].value_counts().nlargest(10).index.tolist()

# Filtering to keep only the top ten DBAs
top10_dba = violations[violations['dba'].isin(top_dba)]

# Computing the average score for each DBA
dba_mean_score = top10_dba.groupby('dba')['score'].mean().reset_index()

# Sorting by mean score in ascending order
mean_score_sorted = dba_mean_score.sort_values(by='score')

mean_score_sorted
```

### STARBUCKS had the lowest average inspection score

## (c)

```python
# Filter for zip codes in Manhattan with 50 or more inspections
manhattan_zips = violations[(violations['boro'] == 'MANHATTAN')]
manhattan_zips = manhattan_zips.groupby('zipcode').filter(lambda x: len(x) >= 50)

# Calculate the median violation score by zip code
median_scores = manhattan_zips.groupby('zipcode')['score'].median().reset_index()

# Print the median scores
median_scores.head()
```

### From observations, there appears to be an inverse relationship between the median score and number of inspections. The zip codes that had more inspections had lower median score. This could be due that the zip codes with more inspections complied with regulations.
