# (a)

## (10 points) Name every player in baseball history who has accumulated at least 300 home runs (HR) AND at least 300 stolen bases (SB). You can find the first and last name of the player in the People data frame. Join this to your result along with the total home runs and total bases stolen for each of these elite players.

```python
player_totals = Batting.loc[:,["playerID","HR","SB"]].groupby("playerID").sum()
players_over300 = player_totals[(player_totals['HR'] >= 300) & (player_totals['SB'] >= 300)]
elite_players = pd.merge(players_over300, People[["playerID", "nameFirst", "nameLast"]], on = "playerID", how = "inner")
elite_players = elite_players[['playerID', 'nameFirst', 'nameLast', 'HR', 'SB']]
elite_players
```

# (c)

## (10 points) Identify the name and year of every player who has hit at least 50 home runs in a single season. Which player had the lowest batting average in that season?


```python
# Identify the name and year of every player who has hit at least 50 home runs in a single season.
part_a = Batting.loc[:,["playerID", "yearID", "HR"]].groupby(["playerID", "yearID"],as_index = False).sum()
part_a = part_a[(part_a["HR"] >= 50)].sort_values("HR",ascending=False)
part_a.head()
```

```python
# join players to People dataset
part_a = pd.merge(part_a, People[["playerID", "nameFirst", "nameLast"]], on="playerID", how = "inner")
part_a.head()
```

```python
# join players to Batting dataset and find the lowest batting average
part_b = pd.merge(part_a, Batting, on=("playerID","yearID"), how="inner")
part_b["batting_avg"]=part_b["H"]/part_b["AB"]
part_c = part_b[["playerID","yearID","nameFirst","nameLast","batting_avg"]].sort_values(by ="batting_avg", ascending = True)
part_c.head()
```

## Mark McGwire had the lowest batting average in that season with a batting average of 0.252874
